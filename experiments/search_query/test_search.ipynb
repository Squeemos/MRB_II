```python
import os
import json
import time

#import google_auth_oauthlib.flow
import googleapiclient.discovery
import googleapiclient.errors

import pandas as pd
```

#### Create API client

```python
with open("../../secret_keys/api_key.txt") as file:
    API_KEY = file.read().strip()

api_service_name = "youtube"
api_version = "v3"

youtube = googleapiclient.discovery.build(
    api_service_name,
    api_version,
    developerKey=API_KEY,
)
```

#### Create and send request

```python
parts = "id"
```

```python
request = youtube.search().list(
        part=parts,
        maxResults=100,
        q="surfing"
)
```

```python
# Run request
response = request.execute()
```

#### Results

```python
print(json.dumps(response, indent=2))
```

```python
len(response["items"])
```

```python
vids = []

for video in response["items"]:
    vids.append(video["id"]["videoId"])
    
vids
```

# Query with Multiple Video IDs

```python
parts = "id,snippet,contentDetails,status,statistics,player,topicDetails,recordingDetails,liveStreamingDetails,localizations"
```

```python
request = youtube.videos().list(
        part=parts,
        id=",".join(vids[:10]),
)
```

```python
# Run request
response = request.execute()
```
